### Part 2

```python
df = pd.read_csv("ecommerce_sales_data.csv")
```

```python
customer_analysis = df.groupby('Customer_ID')['Total_Sales'].sum().sort_values(ascending=False)
```

```python
print("Top 10 Customers by Sale:")
customer_analysis.head(10)
```

```python
customer_frequency = df['Customer_ID'].value_counts()
```

```python
print("\nTop 10 Customers by Purchase Frequency:")
customer_frequency.head(10)
```

### Part 3

```python
product_sales = df.groupby('Product_Name')['Total_Sales'].sum().sort_values(ascending=False)
```

```python
print("Top 10 Products by Sales:")
product_sales.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
top_products = product_sales.head(10)
```

```python
top_products.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title("Top 10 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

### Part 4

```python
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
```

```python
df['Month'] = df['Purchase Date'].dt.to_period('M')
```

```python
df.to_csv("temp.csv")
```

```python
monthly_sales = df.groupby('Month')['Total_Sales'].sum()
```

```python
monthly_sales.plot(kind='line', figsize=(10, 6), marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

### Part 5

```python
df = pd.read_csv("amazonSalesReport.csv")
```

```python
region_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False)
print("Sales by Region:")
print(region_sales)
```

```python
region_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), startangle=90, colors=['skyblue', 'orange', 'lightgreen'])
plt.title("Sales Distribution by Region")
plt.ylabel("")
plt.show()
```
